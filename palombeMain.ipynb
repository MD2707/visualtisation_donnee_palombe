### NOMBRE PALOMBE PAR COL PAR ANNEE

```python
def graphique(mois1, jour1, mois2 ,jour2):
    df = pd.read_csv("finalData.csv")
    
    dfBonneDate1 = df.query("jour >="+str(jour1)+"and mois=="+str(mois1))
    dfBonneDate2 = df.query("jour >="+str(jour2)+"and mois=="+str(mois2))
    dfBonneDate = pd.concat([dfBonneDate1,dfBonneDate2], ignore_index=True)
    listAnnee = []
    listPalombes = []
    totPalombes = 0
    for annee in range(1999,2021):
        listAnnee.append(str(annee))
        dfAnnee = dfBonneDate.query("Annee == "+str(annee))
        dfvalPalombe = dfAnnee["TotalJ"].sum()
        totPalombes+=dfvalPalombe
        listPalombes.append(dfvalPalombe)
        annee+=1
        
    dfFinal=pd.DataFrame({"Annee":listAnnee,"Palombes":listPalombes})
    fig = px.bar(dfFinal, x='Annee', y='Palombes' ,color="Palombes")
    return fig
    
def getMeanNumberOfPalombes(mois1, jour1, mois2 ,jour2):
    df = pd.read_csv("finalData.csv")
    dfBonneDate1 = df.query("jour >="+str(jour1)+"and mois=="+str(mois1))
    dfBonneDate2 = df.query("jour >="+str(jour2)+"and mois=="+str(mois2))
    dfBonneDate = pd.concat([dfBonneDate1,dfBonneDate2], ignore_index=True)
    listAnnee = []
    listPalombes = []
    totPalombes = 0
    for annee in range(1999,2021):
        listAnnee.append(str(annee))
        dfAnnee = dfBonneDate.query("Annee == "+str(annee))
        dfvalPalombe = dfAnnee["TotalJ"].sum()
        totPalombes+=dfvalPalombe
        annee+=1
    return int(totPalombes/(annee-1999))
```
